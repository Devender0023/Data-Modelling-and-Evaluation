```python
# K Nearest neighbour Classifier started 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
y = heart.DEATH_EVENT
data = pd.DataFrame(heart, columns=['age','ejection_fraction','high_blood_pressure','serum_creatinine','serum_sodium','time'])
X_train, X_test, y_train, y_test = train_test_split(data,y,test_size=0.3,random_state=0)
```

```python
data
```

```python
X_train.shape
```

```python
X_train
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
clf = KNeighborsClassifier(5)
fit = clf.fit(X_train, y_train)
predicted = fit.predict(X_test)
print(predicted)
```

```python
predicted.shape
```

```python
cm = confusion_matrix(y_test, predicted)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(clf.score(X_test, y_test))
print(classification_report(y_test,predicted))
```

```python
# Decision Tree Classifier started
```

```python
from sklearn.tree import DecisionTreeClassifier
a_train, a_test, b_train, b_test = train_test_split(data,y,random_state=0)
```

```python
a_train.shape
```

```python
b_train.shape
```

```python
clf = DecisionTreeClassifier()
fit = clf.fit(a_train, b_train)
predicted = fit.predict(a_test)
cm = confusion_matrix(b_test, predicted)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(clf.score(a_test, b_test))
print(classification_report(b_test,predicted))
```
